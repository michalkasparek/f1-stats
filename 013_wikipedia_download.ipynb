```python
import os
import pandas as pd
```

```python
wikidata_ids = pd.read_csv(os.path.join('data','wikidata_ids.csv'))
```

```python
wikidata_ids
```

```python
folder = "data_raw/wikipedia"
```

```python
if not os.path.exists(folder):
    os.makedirs(folder)
    print("Path created successfully!")
else:
    print("Path already exists.")
```

```python
import requests
```

```python
count = 1
total = len(wikidata_ids['driverUrl'].to_list())
for index, row in wikidata_ids.iterrows():
    response = requests.get(row['driverUrl'])
    print(f"{count}/{total} {row['driverUrl']}")
    count = count + 1
    with open(os.path.join(folder, row['wikidata_id'] + '.html'), 'w+', encoding='utf-8') as working_html:
        working_html.write(response.text)
```
