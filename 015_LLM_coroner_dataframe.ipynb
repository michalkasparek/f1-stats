```python
import os
import pandas as pd
```

```python
wikid = pd.read_json(os.path.join('data','wikidata.json'))
```

```python
folder = "data_raw/mistral_deaths"
```

```python
scraped_files = os.listdir(folder)
scraped_qs = []
for f in scraped_files:
    short = f.split('.')[0]
    scraped_qs.append(short)
```

```python
causes_of_death = []
```

```python
for q in scraped_qs:
    dead_driver = {}
    dead_driver['wikidata_id'] = q
    with open(os.path.join(folder,f"{q}.txt"), "r", encoding="utf-8") as file:
        dead_driver['cause_of_death_full'] = file.read()
    causes_of_death.append(dead_driver)
```

```python
causes_of_death
```

```python
df = pd.DataFrame(causes_of_death)
```

```python
len(df)
```

```python
df = df.merge(wikid[['wikidata_id','full_name','date_of_birth','place_of_death','date_of_death','cause_of_death','manner_of_death']], on='wikidata_id', how='inner')
```

```python
df
```

```python
def get_the_cause(mistral):
    cause = None
    causes = ['#gp','#otherf1','#otherracing','#othersport','#traffic','#aviation','#suicide','#violent','#natural','#unknown']
    for c in causes:
        if  c in mistral:
            cause = c.replace('#','')
    return cause
```

```python
df['cause_of_death_mistral'] = df['cause_of_death_full'].apply(lambda x: get_the_cause(x))
```

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
df['date_of_death'] = pd.to_datetime(df['date_of_death'])
```

```python
df['age'] = df['date_of_death'] - df['date_of_birth']
```
