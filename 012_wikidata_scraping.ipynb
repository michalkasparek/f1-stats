```python
def list_of_jsons(slozka):
    seznam = []
    for file in os.listdir(slozka):
        if ".json" in file:
            seznam.append(file)
    return(seznam)
```

```python
def drivers_details(source):
    
    try:

        def get_value(source, keys):
            try:
                for key in keys:
                    source = source[key]
                return source
            except (KeyError, IndexError):
                return None

        driver = {
            'wikidata_id': source,
            'full_name': get_value(source, ['labels', 'en', 'value']),
            'bio': get_value(source, ['descriptions', 'en', 'value']),
            'gender': get_value(source, ['claims', 'P21', 0, 'mainsnak', 'datavalue', 'value', 'id']),
            'date_of_birth': get_value(source, ['claims', 'P569', 0, 'mainsnak', 'datavalue', 'value', 'time']),
            'place_of_birth': get_value(source, ['claims', 'P19', 0, 'mainsnak', 'datavalue', 'value', 'id']),
            'date_of_death': get_value(source, ['claims', 'P570', 0, 'mainsnak', 'datavalue', 'value', 'time']),
            'place_of_death': get_value(source, ['claims', 'P20', 0, 'mainsnak', 'datavalue', 'value', 'id']),
            'cause_of_death': get_value(source, ['claims', 'P509', 0, 'mainsnak', 'datavalue', 'value', 'id']),
            'manner_of_death': get_value(source, ['claims', 'P1196', 0, 'mainsnak', 'datavalue', 'value', 'id']),
            'wikidata_id': get_value(source, ['id'])
        }
        return driver
    
    except:
        return {}
```

```python
wikidata = []
for j in list_of_jsons(folder):
    print(j)
    with open(os.path.join(folder, j), 'r', encoding='utf-8') as working_json:
        wdata = json.load(working_json)
        driver = drivers_details(wdata)
        wikidata.append(driver)
```

```python
try:
    with open(os.path.join("data", 'wikidata_decoded.json'), 'r') as wikidata_decoded_archive:
        already_decoded = json.loads(wikidata_decoded_archive.read())
except:
    already_decoded = {}
already_decoded_keys = list(already_decoded.keys())
```

```python
already_decoded_keys
```

```python
to_decode = []
for x in wikidata:
    for key, val in x.items():
        if val != None:
            if key != 'wikidata_id':
                if val[0] == 'Q':
                    if val not in to_decode:
                        if val not in already_decoded:
                            to_decode.append(val)
```

```python
to_decode
```

```python
import requests
```

```python
def decode(prop):
    url = f'https://www.wikidata.org/wiki/Special:EntityData/{prop}.json'
    response = requests.get(url)
    data = response.json()
    return data['entities'][prop]['labels']['en']['value']
```

```python
decoded = already_decoded
for t in to_decode:
    decoded[t] = decode(t)
```

```python
decoded
```

```python
with open(os.path.join("data", 'wikidata_decoded.json'), 'w+') as wikidata_decoded_archive:
    json.dump(decoded, wikidata_decoded_archive)
```

```python
wikidata_frame = pd.DataFrame(wikidata)
```

```python
wikidata_frame = wikidata_frame.replace(decoded)
```

```python
wikidata_frame['date_of_birth'] = pd.to_datetime(wikidata_frame['date_of_birth'].apply(lambda x: x[1:11] if x else None), errors='coerce')
```

```python
wikidata_frame['date_of_death'] = pd.to_datetime(wikidata_frame['date_of_death'].apply(lambda x: x[1:11] if x else None), errors='coerce')
```
