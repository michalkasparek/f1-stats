```python
import random
from bs4 import BeautifulSoup
```

```python
def get_bio(scraped_file):
    with open(os.path.join('data_raw/wikipedia', scraped_file), encoding='utf-8') as file:
        bio = BeautifulSoup(file.read(), "html.parser")
        bio = bio.find(id='bodyContent')
        bio = '\n'.join([p.text for p in bio.find_all('p')])
        bio = bio.split('Races in bold indicate pole position')[0].split("(key)")[0].strip()
        if len(bio) > 8000:
            bio = bio[0:3990] + "\n\n(...)\n\n" + bio[-3990:]
    return bio
```

## First, we'll try small local models

```python
import ollama
```

```python
def get_cause(biography):
    sys_prompt = f"""You are a research assistant. I will show you a biography of a person and you will pick the cause of the death of the person, based purely on the material I will provide you. You will pick one of these causes of death and you will not add anything else to your answer. KEEP THE HASHTAG IN THE BEGGINING!
    - "#gp died in an race car accident during an F1 Grand Prix",
    - "#otherf1 died in a F1 car but not during a Grand Prix" (for example during testing),
    - "#otherracing died in a race car that was not an F1 car" (for example during Indy 500 or F2 races),
    - "#othersport died doing a sport that wasn't car racing" (for example in a power boats race,
    - "#traffic died in a traffic incident on public roads",
    - "#suicide suicide",
    - "#violent violent death" (for example murder or mugging),
    - "#health bad health", (for example cancer, heart attack or old age)
    - "#unknown the cause of death is unknown or can't be found in the provided material" """
    resp = ollama.generate(model='mistral:latest', prompt=biography, system=sys_prompt, options={'temperature': 0.1})
    return resp['response']
```

## Le inspecteur des cadavres

```python
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage
```

```python
with open(os.path.join('src','mistral.txt')) as mis:
    api_key = mis.read()
```

```python
def le_cause(biography):
    client = MistralClient(api_key=api_key)
    chat_response = client.chat(
    model="mistral-medium-latest",
    temperature=0.1,
    messages= [
        ChatMessage(role="system", content=f"""You are a research assistant. I will show you a truncated biography of a race driver and you will pick the cause of the death of the person, based purely on the material I will provide you. You will pick one of the following causes of death. You can give me one or two extra sentences with additional info or reasoning but at all cost KEEP THE HASHTAG IN THE BEGGINING. Also remember that it is highly unlikely to die in a race after the age of 50.
        The causes of death you are choosing from:
        - "#gp died in an race car accident during an F1 Grand Prix",
        - "#otherf1 died in a F1 car but not during a Grand Prix" (for example during testing),
        - "#otherracing died in a race car that was not an F1 car" (for example during Indy 500 or F2 races),
        - "#othersport died doing a sport that wasn't car racing" (for example in a power boats race or skiing ),
        - "#traffic died in a traffic incident on public roads",
        - "#aviation died in an aviation accident or incident"
        - "#suicide suicide",
        - "#violent violent death" (for example murder or mugging),
        - "#natural natural causes or bad health", (for example cancer, heart attack or old age)
        - "#unknown the cause of death is unknown or can't be found in the provided material" """),
        ChatMessage(role="user", content=biography)
    ])
    print(f"""{biography[0:200]}""")
    return chat_response.choices[0].message.content.strip()
```

```python
out_folder = "data_raw/mistral_deaths"
```

```python
if not os.path.exists(out_folder):
    os.makedirs(out_folder)
    print("Path created successfully!")
else:
    print("Path already exists.")
```

```python
scraped_files = os.listdir(out_folder)
scraped_qs = []
for f in scraped_files:
    short = f.split('.')[0]
    scraped_qs.append(short)
```

```python
for q in dead_qs:
    if q not in scraped_qs:
        b = get_bio(f"{q}.html")
        cause = le_cause(b)
        with open(os.path.join(out_folder,f"{q}.txt"), "w+", encoding="utf-8") as output_file:
            output_file.write(cause)
```
