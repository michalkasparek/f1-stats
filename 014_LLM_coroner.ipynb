## First, we'll try small local models

```python
import ollama
```

```python
def get_cause(biography):
    sys_prompt = f"""You are a research assistant. I will show you a biography of a person and you will pick the cause of the death of the person, based purely on the material I will provide you. You will pick one of these causes of death and you will not add anything else to your answer. KEEP THE HASHTAG IN THE BEGGINING!
    - "#gp died in an race car accident during an F1 Grand Prix",
    - "#otherf1 died in a F1 car but not during a Grand Prix" (for example during testing),
    - "#otherracing died in a race car that was not an F1 car" (for example during Indy 500 or F2 races),
    - "#othersport died doing a sport that wasn't car racing" (for example in a power boats race,
    - "#traffic died in a traffic incident on public roads",
    - "#suicide suicide",
    - "#violent violent death" (for example murder or mugging),
    - "#health bad health", (for example cancer, heart attack or old age)
    - "#unknown the cause of death is unknown or can't be found in the provided material" """
    resp = ollama.generate(model='mistral:latest', prompt=biography, system=sys_prompt, options={'temperature': 0.1})
    return resp['response']
```

## Le vérificateur de cadavres

```python
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage
```

```python
with open(os.path.join('src','mistral.txt')) as mis:
    api_key = mis.read()
```

```python
model = "mistral-medium-latest"
```

```python
client = MistralClient(api_key=api_key)
```

```python
chat_response = client.chat(
    model=model,
    temperature=0.1,
    messages= [
    ChatMessage(role="system", content=f"""You are a research assistant. I will show you a biography of a person and you will pick the cause of the death of the person, based purely on the material I will provide you. You will pick one of the following causes of death. You can give me one or two extra sentences with additional info or reasoning but at all cost KEEP THE HASHTAG IN THE BEGGININ!
    - "#gp died in an race car accident during an F1 Grand Prix",
    - "#otherf1 died in a F1 car but not during a Grand Prix" (for example during testing),
    - "#otherracing died in a race car that was not an F1 car" (for example during Indy 500 or F2 races),
    - "#othersport died doing a sport that wasn't car racing" (for example in a power boats race or skiing ),
    - "#traffic died in a traffic incident on public roads",
    - "#aviation died in an aviation accident or incident"
    - "#suicide suicide",
    - "#violent violent death" (for example murder or mugging),
    - "#health bad health", (for example cancer, heart attack or old age)
    - "#unknown the cause of death is unknown or can't be found in the provided material" """),
ChatMessage(role="user", content=bio)
]
)
```
