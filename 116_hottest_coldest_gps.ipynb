```python
import polars as pl

df = pl.read_parquet('data/gps_wikipedia.parquet')
```

```python
df.select(pl.col('Weather')).sample(10)
```

```python
test = df.select(pl.col('Weather')).unique().sample(50).to_series().to_list()
```

```python
import re
```

```python
def temperature(weat):
    if "°C" not in weat:
        return None
    else:
        weat = weat.split(" °C")[0].split(" ")[-1]
        return weat
```

```python
for t in test:
    print(t)
    print(temperature(t))
```
