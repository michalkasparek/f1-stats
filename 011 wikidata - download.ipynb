```python
folder = "data/wikidata"
```

```python
if not os.path.exists(folder):
    os.makedirs(folder)
    print("Path created successfully!")
else:
    print("Path already exists.")
```

```python
from wikidata.client import Client
import wikidata.entity
```

```python
def wikidates(id):
    client = Client()
    entity = client.get(id, load=True)
    return entity
```

```python
import json
```

```python
count = 1
total = len(wikidata_ids['wikidata_id'].to_list())
for q in wikidata_ids['wikidata_id'].to_list():
    print(f"{count}/{total}")
    count = count + 1
    with open(os.path.join(folder, q + '.json'), 'w+', encoding='utf-8') as working_json:
        json_data = json.dumps(wikidates(q).data)
        working_json.write(json_data)
```
