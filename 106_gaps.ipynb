```python
import os
import pandas as pd

results = pd.read_json(os.path.join('data','ergast-results.json'))
```

```python
def vteriny(x):
    try:
        x = str(x).replace('+','')
        if ':' in x:
            mins = x.split(':')[0]
            secs = x.split(':')[1]
            x = int(mins) + float(secs)
        return float(x)
    except:
        return None
```

```python
secondplaces = results[results['position'] == 2]
secondplaces['time'] = secondplaces['time'].apply(lambda x: vteriny(x))
```
