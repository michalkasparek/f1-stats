```python
import pandas as pd
df = pd.read_json("data/ergast-results.json")
```

```python
df.columns
```

```python
first_wins = df[df['position'] == 1].sort_values(by=["date"])[['date','constructor']].drop_duplicates(subset=['constructor'],keep='first')
first_wins
```

```python
import polars as pl
import altair as alt
```

```python
fw_pl = pl.from_pandas(first_wins).group_by_dynamic(index_column="date", every="1y").agg(pl.col("constructor").unique().len())
fw_pl
```

```python
alt.Chart(pd.DataFrame([{"a":1,"b":2}])).mark_bar().encode(alt.X("a:T"),alt.Y("b:Q"))
```
