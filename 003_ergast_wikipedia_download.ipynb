```python
import os
import requests
import pandas as pd
races = pd.read_csv("data_raw/ergast/races.csv")
```

```python
urls = races['url'].to_list()
```

```python
len(urls)
```

```python
dowl_loc = "data_raw/downloads/wikipedia/races"

if not os.path.exists(dowl_loc):
   os.makedirs(dowl_loc)
```

```python
downloaded = os.listdir(dowl_loc)
```

```python
for u in urls:
    filename = u.split('/')[-1] + ".html"
    if filename not in downloaded:
        r = requests.get(u, stream=True)
        if r.status_code == 200:
            with open(os.path.join(dowl_loc, filename), 'w+', encoding='utf-8') as f:
                f.write(r.text)
    print(filename)
```
