```python
df = pd.read_json("data/ergast-results.json")
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['age'] = df['date'] - df['dob']
```

```python
forty = 40*365.25
```

```python
count = df[df['age'].dt.days >= forty].groupby(['year','gp']).size()
count[count >= 2].tail(30)
```
