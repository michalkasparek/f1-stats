```python
from calendar import monthrange
```

```python
def monthday(date_value):
    # Here date_value is a datetime.datetime object, not a Polars column
    month_length = monthrange(date_value.year, date_value.month)[1]
    return date_value.month + (date_value.day / month_length)
```

```python
df = df.with_columns(pl.col("date").dt.ordinal_day().alias("day"))
df = df.with_columns(pl.col("date").dt.week().alias("week"))
df = df.with_columns(pl.col("date").map_elements(monthday).alias("monthday"))

asia = ['UAE','India','Bahrain','Malaysia', 'Qatar', 'China','Japan','Singapore',
             'Saudi Arabia','Korea','Azerbaijan','Turkey']
america = ['USA','Brazil', 'Canada', 'Argentina','Mexico','United States']
europe = ['Netherlands','Portugal','Germany','UK','France','Belgium','Sweden','Switzerland','Italy','Spain','Russia','Monaco','Hungary','Austria']
oceania = ['Australia']
africa = ['South Africa','Morocco']

street = ['Nevada','Las Vegas','Jeddah','Dallas','Montreal','Detroit','Phoenix','Monte-Carlo','Oporto','Adelaide','Baku','Marina Bay','Valencia','Miami']

nightIds = df.filter(pl.col("location").is_in(['Las Vegas','Jeddah','Marina Bay','Abu Dhabi','Al Daayen']) | ((pl.col('country') == 'Bahrain') & (pl.col('year') >= 2014))).select(pl.col('raceId')).to_series().to_list()

def get_continent(country): ## this is the ugliest thing ever writtern but I'm being sleepy and curious at the same time
    if country in asia:
        return "asia"
    if country in america:
        return "america"
    if country in europe:
        return "europe"
    if country in oceania:
        return "oceania"
    if country in africa:
        return "africa"

def get_type(location):
    if location in street:
        return "street"
    else:
        return "road"

def get_light(id):
    if id in nightIds:
        return "night"
    else:
        return "day"

df = df.with_columns(pl.col('country').map_elements(get_continent).alias('continent'))
df = df.with_columns(pl.col('location').map_elements(get_type).alias('type'))
df = df.with_columns(pl.col('raceId').map_elements(get_light).alias('day_night'))
```

```python
championship_races = df.group_by("raceId").agg(pl.col(["date","year","week","day","monthday","continent","type","day_night","location"]).min()).join(
    df.group_by("raceId").agg(pl.col('constructor').unique().len().alias('constructors')), how='left', on='raceId').join(
    df.group_by("raceId").agg(pl.col('driverId').unique().len().alias('drivers')), how='left', on='raceId').join(
    pl.read_parquet('data/gps_wikipedia.parquet'), on='raceId', how='left'
    ).with_columns(
    (pl.col("Kms") / pl.col("Laps")).alias("Course lenght")
    ).with_columns(
    pl.col("year").map_elements(lambda x: str(x)[0:3]).alias('decade')
    )

championship_races
```

```python
legenda = "right"

alt.Chart(
    championship_races.to_pandas(),
    width=360,
    title=alt.Title("75 let MS F1: proměny kalendáře")
).mark_point(filled=True
).encode(
    alt.Y("year:O", axis=alt.Axis(domainOpacity=0, tickColor='#DCDDD6'), title=None),
    alt.X("monthday:Q", axis=alt.Axis(domainOpacity=0, tickColor='#DCDDD6'), scale=alt.Scale(domainMax=12.9, domainMin=1), title=None),
    alt.Color('continent:N', legend=alt.Legend(orient=legenda, direction="vertical")),
    alt.Size('Course lenght:Q', legend=alt.Legend(orient=legenda, direction="vertical")),
    alt.Shape('type:N', scale=alt.Scale(range=['circle','diamond']), legend=alt.Legend(orient=legenda, direction="vertical")),
    alt.Stroke('day_night:N', 
               scale=alt.Scale(domain=['day', 'night'], range=['white', 'black']), legend=alt.Legend(orient=legenda, direction="vertical")),
    alt.StrokeWidth('day_night:N',
                   scale=alt.Scale(domain=['day', 'night'], range=[0, 1.5]), legend=alt.Legend(orient=legenda, direction="vertical")),
    alt.Row('decade:O', title=None)
).resolve_axis(y='independent',x='independent').resolve_scale(y='independent',color='shared').configure_view(
    stroke='transparent'
)
```
