```python
import os
import pandas as pd
```

```python
results = pd.read_json(os.path.join('data','ergast-results.json'))
```

```python
results.groupby(['date','constructor','name'])['position'].first()
```

```python
results = results.sort_values(by="position")
```

```python
comp = pd.DataFrame(results.groupby(['date','constructor','name'])['position'].first())
comp
```

```python
comp['best_in_team'] = comp.groupby(['date', 'constructor'])['position'].transform(lambda x: x == x.min())
```

```python
comp
```

```python
average_result = pd.DataFrame(comp.groupby('name')['position'].mean())
average_result = average_result = average_result.reset_index()
average_result = average_result.rename(columns={'position':'avg','name':'driver'})
```

```python
comp['driver'] = comp.index.get_level_values(2)
```

```python
comp
```

```python
average_result
```

```python
comp = comp.reset_index(drop=False).merge(average_result, on='driver')
```

```python
comp
```

```python
comp
```

```python
comp[comp['best_in_team'] == True].groupby('driver')['avg'].mean().sort_values().head(50)
```

```python
comp['drivers_beaten'] = comp.groupby(['date', 'constructor'])['avg'].min()
```

```python
comp
```

```python
comp
```

```python
best_in_team_counts = comp[comp['best_in_team'] == True]['name'].value_counts()
```

```python
total_races_counts = comp['name'].value_counts()
```

```python
total_races_counts
```

```python
total_races_counts = total_races_counts[total_races_counts >= 40]
```

```python
drivers_percentages = (best_in_team_counts / total_races_counts) * 100
```

```python
drivers_percentages.dropna().sort_values(ascending=False).head(30)
```
