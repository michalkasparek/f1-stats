```python
import os
import pandas as pd

results = pd.read_json(os.path.join('data','ergast-results.json'))
```

```python
debuts = pd.DataFrame(results.groupby('name')['date'].min()).reset_index()
```

```python
debuts.sort_values(by="date",ascending=False).head(30)
```

```python
debuts_cons = debuts.merge(results[['name','date','constructor','gp']], on=['name','date'], how='inner')
```

```python
debuts_cons[debuts_cons['constructor'] == 'Ferrari'].sort_values(by='date', ascending=False)
```
